```python
from kafka import KafkaConsumer
import json
import numpy as np
```

```python
consumer=KafkaConsumer('lavapies',bootstrap_servers='192.168.1.71:9092')
```

```python
txt=""
for msg in consumer:
    #Recupera en formato json cada post
    mi_json = json.loads(msg.value.decode('utf8'))
    

    txt += mi_json['text']
    print('==============')
    print(mi_json['text'])
    print('==============')
    print()
    print()
```

```python
file = open('tweets.txt','w',encoding="utf-8")
for linea in txt:    
    file.write(linea)
file.close()
```

```python
import findspark
findspark.init()

import pyspark
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName('appx')
sc = SparkContext(conf = conf)
```

```python
contentRDD = sc.textFile('tweets.txt')
```

```python
nonempty_lines = contentRDD.filter(lambda x: len(x) > 0)
```

```python
words = nonempty_lines.flatMap(lambda x: x.split(' '))
```

```python
wordcount = words.map(lambda x:(x,1)).reduceByKey(lambda x,y: x+y).map(lambda x: (x[1], x[0])).sortByKey(False)
```

```python
wc = open('wordcount.txt','w',encoding="utf-8")
for word in wordcount.collect():
    wc.write(str(word[0])+"|"+str(word[1])+';\n')
wc.close()
```
